# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
# Your code here
# Set the value of n to 20
n = 20

# Print the table headers
print("r\tV")

# Iterate through numbers from 1 to 20 (inclusive)
for i in range(1, n + 1):
    # Check if the current number (i) is even
    if i % 2 == 0:
        # Assign the current number to the radius (r)
        r = i

        # Calculate the volume (V) using the formula for the volume of a sphere
        V = (2 / 3) * (3.14) * (r ** 3)

        # Print the radius and corresponding volume in a formatted way
        print("{}\t{:.2f}".format(r, V))

```

# Question 2 (20 points)

(a) Write a Python program to generate a sequence of 12 random numbers falling within the range of 50 to 150.

(b) Arrange the generated numbers in ascending order. Display the sorted list.

(c) Extend the program to arrange the numbers in descending order as well. Present the sorted list accordingly.

(d) Compute and output the average of the entire group of generated numbers.

(e) Enhance the program to calculate and display the average of the first 6 numbers from both the ascending and descending sorted lists.

```python
# Import the random module to generate random numbers
import random

# Generate a list of 12 random numbers within the range of 50 to 150
random_numbers = [random.randint(50, 150) for i in range(12)]

# Sort the generated random numbers in ascending order
random_numbers_asc = sorted(random_numbers)

# Sort the generated random numbers in descending order
random_numbers_des = sorted(random_numbers, reverse=True)

# Calculate the average of the generated random numbers
avg_random_numbers = sum(random_numbers) / len(random_numbers)

# Print the generated random numbers, ascending order, descending order, and average
print('Generated random 12 numbers falling within the range of 50 to 150:', random_numbers)
print("The Ascending order of generated numbers is: ", random_numbers_asc)
print("The Descending order of generated numbers is: ", random_numbers_des)
print("The Average of generated numbers is: ", round(avg_random_numbers, 2))

# Calculate the sum of the first 6 numbers from both ascending and descending sorted lists
sum = 0
for i in range(6):
    sum = sum + random_numbers_asc[i] + random_numbers_des[i]

# Calculate the average of the sums
average_of_sums = sum / 12

# Print the average of the sums
print("The average of the first 6 numbers from both the ascending and descending sorted lists is: ", average_of_sums)
```

# Question 4 (20 Points)

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(c) Enhance your calculator to include additional mathematical operations such as exponentiation, square root, and modulus.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
# Your code here
import re
import math

def evaluate_expression(expression):
    try:
        # Replace "^" with "**" for exponentiation
        expression = re.sub(r'\^', r'**', expression)

        # Replace "sqrt" with "math.sqrt" for square root
        expression = re.sub(r'sqrt\(', r'math.sqrt(', expression)

        # Evaluate the expression
        result = eval(expression)

        return result
    except ZeroDivisionError:
        return "Error: Division by zero"
    except Exception as e:
        return f"Error: {str(e)}"

def calculate():
    print("Command-Line Calculator")
    print("Type 'exit' to quit.")

    while True:
        user_input = input("Enter a mathematical expression: ")

        if user_input.lower() == 'exit':
            print("Exiting calculator. Goodbye!")
            break

        # Validate the input expression
        if not re.match(r'^[0-9+\-*/().\s^sqrt%]+$', user_input):
            print("Invalid input. Please enter a valid mathematical expression.")
            continue

        result = evaluate_expression(user_input)
        print("Result:", result)
calculate()
```

The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----
	2012 ----- -----
    2013 ----- -----
	2014 ----- -----
  
  

(c) (10 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (10 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
 # Your code here
import pandas as pd
data=pd.read_csv('/content/Assignment1_denton_housing.csv')

# Filtering the DataFrame for 'Occupied Housing Units' title_field
data_occupied = data[data['title_field'] == 'Occupied Housing Units']

# Sorting the DataFrame by 'year' for consecutive calculations
data_occupied = data_occupied.sort_values(by='year')

# Update the 'occupied_Housing_Units_Difference' column with the differences in the 'Value' column
data_occupied['occupied_Housing_Units_Difference'] = data_occupied['value'].diff()

# setting the first row of 'occupied_Housing_Units_Difference' to zero
data_occupied.loc[data_occupied.index[0], 'occupied_Housing_Units_Difference'] = 0

# Displaying the calculated difference
print(data_occupied[['year', 'value', 'occupied_Housing_Units_Difference']])





```

```python
#  Filter the DataFrame for 'Vacant Housing Units' and 'Occupied Housing Units'
data_vacant = data[data['title_field'] == 'Vacant Housing Units']
data_occupied = data[data['title_field'] == 'Occupied Housing Units']
data_total=data[data['title_field']=='Total Housing Units']

merged_data = pd.merge(data_vacant[['year', 'value']], data_occupied[['year', 'value']], on='year',suffixes=('_vacant', '_occupied'))
merged_data=pd.merge(merged_data,data_total[['year','value']],on='year',)
merged_data=merged_data.sort_values(by='year')
for index, row in merged_data.iterrows():
    year = row['year']
    vacant_units = row['value_vacant']
    occupied_units = row['value_occupied']
    total_units = row['value']
    vacant_percentage = (vacant_units / total_units) * 100
    occupied_percentage = (occupied_units / total_units) * 100
    print(f"{year} {vacant_percentage:.0f}% {occupied_percentage:.0f}%")
```

```python
# Sorting the DataFrames in descending order by 'Value'
data_vacant=data_vacant.sort_values(by='value',ascending=False)
data_occupied=data_occupied.sort_values(by='value',ascending=False)

# Finding the year with the highest number of vacant housing units
max_vacant_year = data_vacant.iloc[0]['year']
max_vacant_value = data_vacant.iloc[0]['value']

# Finding the year with the highest number of occupied housing units
max_occupied_year = data_occupied.iloc[0]['year']
max_occupied_value = data_occupied.iloc[0]['value']

df = pd.DataFrame({

     'Year': [max_vacant_year, max_occupied_year],
    'Value': [max_vacant_value, max_occupied_value],

})

# Adjust values for clarity
df.index  = ["Vacant Housing Units ", "Occupied Housing Units"]

df
```

```python
import statistics

# Filter the DataFrame for total housing units
data_total = data[data['title_field'] == 'Total Housing Units']

# Calculate the harmonic mean using the statistics module
harmonic_mean_value = statistics.harmonic_mean(data_total['value'])

# Print the harmonic mean
print(f"The harmonic mean of total housing units is: {harmonic_mean_value}")
```
